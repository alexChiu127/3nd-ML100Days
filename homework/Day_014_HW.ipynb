```python
import numpy as np
np.random.seed(1)

import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

### 弱相關

```python
x = np.random.randint(0, 50 ,1000)
y = np.random.randint(0, 50 ,1000)

np.corrcoef(x,y)
```

```python
plt.scatter(x, y)
```

### 正相關

```python
normal = np.random.normal(0,10,1000)
plt.plot(normal)
```

```python
x = np.random.randint(0, 50, 1000)
y = x + np.random.normal(0,10,1000)
```

```python
plt.scatter(x, y)
```
